```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read in CSVs
df1 = pd.read_csv("Resources/OSMI 2019 Mental Health in Tech Survey Results.csv")
df2 = pd.read_csv("Resources/OSMI 2020 Mental Health in Tech Survey Results cleaned.csv")
df3 = pd.read_csv("Resources/OSMI Mental Health in Tech Survey 2018.csv")
df4 = pd.read_csv("Resources/OSMI Mental Health in Tech Survey 2017.csv")
```

```python
# Merge CSVs
final_df = pd.concat([df1,df2,df3,df4]).reset_index(drop=True)
final_df = final_df.dropna()
final_df.to_csv("mergedmentalhealth.csv",index=False)
final_df
```

```python
age_average = final_df.groupby("year")["age"].mean().reset_index()
age_average
age_difficulty = final_df.groupby(["year","difficulty_leave"]).count().reset_index()
age_difficulty
```

```python
# Load dataframes into databases
final_df.to_sql(name = 'mental_health', con = engine, if_exists = 'append', index = False)
```
